```python
# 📦 Importing all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
```

Dataset loading


```python
df = pd.read_csv('/content/heart.csv')

print('✅ Dataset loaded successfully!')
print(f'📊 Shape: {df.shape[0]} rows × {df.shape[1]} columns')
```

Data Cleaning


```python
# Null Values
print('Missing values per column:')
print(df.isnull().sum())
print(f'\nTotal missing values: {df.isnull().sum().sum()}')
```

```python
# Duplicate values
duplicates = df.duplicated().sum()
print(f'Number of duplicate rows: {duplicates}')
```

```python
# Droping Duplicate value
if duplicates > 0:
    df = df.drop_duplicates()
    print(f'Duplicates removed! New shape: {df.shape}')
else:
    print('No duplicates found!')
```

```python
print('Target column value counts:')
print(df['target'].value_counts())
print(f'\n0 = No Heart Disease')
print(f'1 = Has Heart Disease')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
# Count Plot
labels_map = {0: 'No Disease', 1: 'Has Disease'}
target_counts = df['target'].value_counts()

axes[0].bar(['No Disease', 'Has Disease'], target_counts.values,
            color=['Green', 'Red'], edgecolor='black', width=0.5)
axes[0].set_title('Heart Disease Count', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Number of Patients')
for i, v in enumerate(target_counts.values):
    axes[0].text(i, v + 2, str(v), ha='center', fontweight='bold', fontsize=12)


# Pie Chart
axes[1].pie(target_counts.values, labels=['No Disease', 'Has Disease'],
            autopct='%1.1f%%', colors=['Green', 'Red'],
            startangle=90, explode=(0.05, 0.05))
axes[1].set_title('Heart Disease Percentage', fontsize=14, fontweight='bold')

plt.suptitle('Target Variable Distribution', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print(f'\n Insight: {target_counts[1]} patients have heart disease ({100*target_counts[1]/len(df):.1f}%)')
print(f'           {target_counts[0]} patients do NOT have heart disease ({100*target_counts[0]/len(df):.1f}%)')
```

```python
# Age distribution by Heart Disease
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# Histogram of age
axes[0].hist(df[df['target'] == 0]['age'], bins=20, alpha=0.7,
             color='Green', label='No Disease', edgecolor='white')
axes[0].hist(df[df['target'] == 1]['age'], bins=20, alpha=0.7,
             color='Red', label='Has Disease', edgecolor='white')
axes[0].set_title('Age Distribution by Heart Disease', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Count')
axes[0].legend()

# Boxplot
df_plot = df.copy()
df_plot['Disease'] = df_plot['target'].map({0: 'No Disease', 1: 'Has Disease'})
sns.boxplot(data=df_plot, x='Disease', y='age',
            palette={'No Disease': 'Green', 'Has Disease': 'Red'}, ax=axes[1])
axes[1].set_title('Age Boxplot by Heart Disease', fontsize=13, fontweight='bold')
axes[1].set_xlabel('')
axes[1].set_ylabel('Age')

plt.tight_layout()
plt.show()

print(f"\n Average age (No Disease)  : {df[df['target']==0]['age'].mean():.1f} years")
print(f" Average age (Has Disease) : {df[df['target']==1]['age'].mean():.1f} years")
```

```python
# 👨👩 Gender vs Heart Disease
df_plot = df.copy()
df_plot['Sex'] = df_plot['sex'].map({1: 'Male', 0: 'Female'})
df_plot['Disease'] = df_plot['target'].map({0: 'No Disease', 1: 'Has Disease'})

fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# Count by gender
gender_counts = df_plot.groupby(['Sex', 'Disease']).size().unstack()
gender_counts.plot(kind='bar', ax=axes[0], color=['lightGreen', 'Red'],
                   edgecolor='black', width=0.6)
axes[0].set_title('Heart Disease by Gender', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=0)
axes[0].legend(title='Condition')

# Pie chart for gender distribution
sex_counts = df_plot['Sex'].value_counts()
axes[1].pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%',
            colors=['skyblue', 'Violet'], startangle=90, explode=(0.05, 0.05))
axes[1].set_title('Gender Distribution in Dataset', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.show()

print('\n Gender breakdown:')
print(df_plot['Sex'].value_counts())
```

```python
#Chest Pain Type vs Heart Disease
cp_map = {0: 'Typical Angina', 1: 'Atypical Angina',
          2: 'Non-Anginal Pain', 3: 'Asymptomatic'}

df_plot = df.copy()
df_plot['CP Type'] = df_plot['cp'].map(cp_map)
df_plot['Disease'] = df_plot['target'].map({0: 'No Disease', 1: 'Has Disease'})

plt.figure(figsize=(8, 4))
cp_disease = df_plot.groupby(['CP Type', 'Disease']).size().unstack(fill_value=0)
cp_disease.plot(kind='bar', color=['lightgreen', 'Red'], edgecolor='black',
                width=0.65, figsize=(12, 5))
plt.title('Chest Pain Type vs Heart Disease', fontsize=14, fontweight='bold')
plt.xlabel('Chest Pain Type')
plt.ylabel('Number of Patients')
plt.xticks(rotation=15)
plt.legend(title='Condition')
plt.tight_layout()
plt.show()

print('\n📌 Chest Pain Types:')
print('  0 = Typical Angina    | Chest pain from reduced blood flow')
print('  1 = Atypical Angina   | Chest pain NOT from heart')
print('  2 = Non-Anginal Pain  | Non-heart-related pain')
print('  3 = Asymptomatic      | No chest pain at all')
```

```python
# Cholesterol & Blood Pressure vs Heart Disease
df_plot = df.copy()
df_plot['Disease'] = df_plot['target'].map({0: 'No Disease', 1: 'Has Disease'})

fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# Cholesterol
sns.violinplot(data=df_plot, x='Disease', y='chol',
               palette={'No Disease': 'lightgreen', 'Has Disease': 'red'},
               ax=axes[0], inner='quartile')
axes[0].set_title('Cholesterol by Heart Disease', fontsize=13, fontweight='bold')
axes[0].set_xlabel('')
axes[0].set_ylabel('Cholesterol (mg/dl)')
axes[0].axhline(y=200, color='orange', linestyle='--', alpha=0.7, label='Borderline (200)')
axes[0].axhline(y=240, color='red', linestyle='--', alpha=0.7, label='High (240)')
axes[0].legend(fontsize=9)

# Resting Blood Pressure
sns.violinplot(data=df_plot, x='Disease', y='trestbps',
               palette={'No Disease': 'lightgreen', 'Has Disease': 'red'},
               ax=axes[1], inner='quartile')
axes[1].set_title('Resting Blood Pressure by Heart Disease', fontsize=13, fontweight='bold')
axes[1].set_xlabel('')
axes[1].set_ylabel('Blood Pressure (mm Hg)')
axes[1].axhline(y=120, color='orange', linestyle='--', alpha=0.7, label='Normal (120)')
axes[1].axhline(y=140, color='red', linestyle='--', alpha=0.7, label='High (140)')
axes[1].legend(fontsize=9)

plt.tight_layout()
plt.show()

print(f"\n📌 Avg Cholesterol (No Disease):  {df[df['target']==0]['chol'].mean():.1f} mg/dl")
print(f"📌 Avg Cholesterol (Has Disease): {df[df['target']==1]['chol'].mean():.1f} mg/dl")
```

```python
# Max Heart Rate vs Age (Scatter Plot)
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

colors = df['target'].map({0: '#2ecc71', 1: '#e74c3c'})

# Scatter plot
axes[0].scatter(df['age'], df['thalach'], c=colors, alpha=0.6, s=60, edgecolors='white')
axes[0].set_title('Age vs Max Heart Rate', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Max Heart Rate (thalach)')
patch1 = mpatches.Patch(color='lightgreen', label='No Disease')
patch2 = mpatches.Patch(color='Red', label='Has Disease')
axes[0].legend(handles=[patch1, patch2])

# Boxplot
df_plot = df.copy()
df_plot['Disease'] = df_plot['target'].map({0: 'No Disease', 1: 'Has Disease'})
sns.boxplot(data=df_plot, x='Disease', y='thalach',
            palette={'No Disease': 'lightgreen', 'Has Disease': 'Red'}, ax=axes[1])
axes[1].set_title('Max Heart Rate by Heart Disease', fontsize=13, fontweight='bold')
axes[1].set_ylabel('Max Heart Rate')
axes[1].set_xlabel('')

plt.tight_layout()
plt.show()

print(f"\n Avg Max Heart Rate (No Disease):  {df[df['target']==0]['thalach'].mean():.1f}")
print(f"Avg Max Heart Rate (Has Disease): {df[df['target']==1]['thalach'].mean():.1f}")
print('\n Higher max heart rate is associated with heart disease presence in this dataset!')
```

```python
# Fasting Blood Sugar vs Heart Disease
df_plot = df.copy()
df_plot['FBS'] = df_plot['fbs'].map({1: 'FBS > 120 mg/dl', 0: 'FBS ≤ 120 mg/dl'})
df_plot['Disease'] = df_plot['target'].map({0: 'No Disease', 1: 'Has Disease'})

fbs_disease = df_plot.groupby(['FBS', 'Disease']).size().unstack(fill_value=0)

fbs_disease.plot(kind='bar', color=['lightgreen', 'Red'], edgecolor='black',
                 width=0.5, figsize=(8, 4))
plt.title('Fasting Blood Sugar vs Heart Disease', fontsize=14, fontweight='bold')
plt.xlabel('Fasting Blood Sugar')
plt.ylabel('Number of Patients')
plt.xticks(rotation=0)
plt.legend(title='Condition')
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap — How all features relate to each other
plt.figure(figsize=(8, 9))
corr_matrix = df.corr(numeric_only=True).round(2)

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))  # Only lower triangle

sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f',
            cmap='RdYlGn', center=0, vmin=-1, vmax=1,
            linewidths=0.5, square=True,
            cbar_kws={'shrink': 0.8})

plt.title('Correlation Heatmap of All Features', fontsize=15, fontweight='bold', pad=15)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Print top correlations with target
print('\n📌 Top features correlated with Heart Disease (target):')
target_corr = corr_matrix['target'].drop('target').sort_values(key=abs, ascending=False)
print(target_corr.to_string())
```

```python
# Boxplots to detect outliers in numeric features
num_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

fig, axes = plt.subplots(1, 5, figsize=(8,4))

for i, col in enumerate(num_features):
    axes[i].boxplot(df[col].dropna(), patch_artist=True,
                    boxprops=dict(facecolor='skyblue', color='Black'),
                    medianprops=dict(color='red', linewidth=2),
                    whiskerprops=dict(color='Black'),
                    flierprops=dict(markerfacecolor='orange', marker='o', markersize=6))
    axes[i].set_title(col.upper(), fontweight='bold')
    axes[i].set_xticks([])

plt.suptitle('Outlier Detection — Boxplots', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Count outliers using IQR method
print('\n📌 Outlier Count (IQR Method):')
for col in num_features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    print(f'  {col:12s}: {len(outliers)} outliers')
```

```python
# Final Summary Statistics
print('=' * 60)
print('        HEART DISEASE ANALYSIS — KEY INSIGHTS')
print('=' * 60)

total = len(df)
disease_pct = df['target'].mean() * 100

print(f'\n Dataset Summary:')
print(f'   Total Patients        : {total}')
print(f'   Has Heart Disease     : {df["target"].sum()} ({disease_pct:.1f}%)')
print(f'   No Heart Disease      : {total - df["target"].sum()} ({100-disease_pct:.1f}%)')

print(f'\n Demographics:')
print(f'   Age Range             : {df["age"].min()} – {df["age"].max()} years')
print(f'   Average Age           : {df["age"].mean():.1f} years')
print(f'   Male Patients         : {df["sex"].sum()} ({100*df["sex"].mean():.1f}%)')
print(f'   Female Patients       : {total - df["sex"].sum()} ({100*(1-df["sex"].mean()):.1f}%)')

print(f'\n Health Metrics (Average):')
print(f'   Resting BP            : {df["trestbps"].mean():.1f} mm Hg')
print(f'   Cholesterol           : {df["chol"].mean():.1f} mg/dl')
print(f'   Max Heart Rate        : {df["thalach"].mean():.1f}')

print(f'\n Strongest Correlations with Heart Disease:')
corr = df.corr(numeric_only=True)['target'].drop('target').sort_values(key=abs, ascending=False)
for feat, val in corr.head(5).items():
    direction = ' Positive' if val > 0 else ' Negative'
    print(f'   {feat:12s}: {val:+.3f}  ({direction})')

print('\n' + '=' * 60)
print(' Analysis Complete')
print('=' * 60)
```
